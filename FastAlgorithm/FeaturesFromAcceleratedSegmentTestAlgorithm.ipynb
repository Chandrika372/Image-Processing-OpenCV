```python
# FAST (Features from Accelerated Segment Test) Algorithm for Corner Detection
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the image
image1 = cv2.imread("C:/Users/HY/Downloads/MyAvatar.png")

# Change BGR to RGB
image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

# display the loaded image
plt.imshow(image)
plt.show()
```

```python
fast = cv2.FastFeatureDetector_create()

# Detect the keypoints with non max suppression
nonmax_keypoints = fast.detect(image)

# Disable NonmaxSuppression
fast.setNonmaxSuppression(False)

# Detect the keypoints without non max suppression
without_nonmax_keypoints = fast.detect(image)

nonmax_image = np.copy(image)
without_nonmax_image = np.copy(image)
```

```python
# Draw keypoints on top of the input image
cv2.drawKeypoints(image, nonmax_keypoints, nonmax_image, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.drawKeypoints(image, without_nonmax_keypoints, without_nonmax_image, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)


# Display image with non max suppression
fx, plots = plt.subplots(1, 2, figsize=(15, 15))

plots[0].set_title("Image with NonMax Suppression")
plots[0].imshow(nonmax_image)

# Number of keypoints detected in the training image
print("Number of Keypoints Detected In The Image With NonMax Suppression: ", len(nonmax_keypoints))

# Display image without non max suppression
plots[1].set_title("Image without NonMax Suppression")
plots[1].imshow(without_nonmax_image)

# Number of keypoints detected in the query image
print("Number of Keypoints Detected In The Image Without NonMax Suppression: ", len(without_nonmax_keypoints))
```
